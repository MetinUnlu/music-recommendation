## Normalizing Play count

```python
# Assuming 'matrix' is your list of lists or list of tuples
matrix=df.to_numpy()

# Initialize a dictionary to store min and max play counts for each user
playcount_stats = {}

# Calculate min and max play counts for each user
for track, user_id, play_count in matrix:
    if user_id not in playcount_stats:
        playcount_stats[user_id] = {'min': play_count, 'max': play_count}
    else:
        playcount_stats[user_id]['min'] = min(play_count, playcount_stats[user_id]['min'])
        playcount_stats[user_id]['max'] = max(play_count, playcount_stats[user_id]['max'])

# Normalize the play counts in the matrix
for i, (music_id, user_id, play_count) in enumerate(matrix):
    min_playcount = playcount_stats[user_id]['min']
    max_playcount = playcount_stats[user_id]['max']
    # Avoid division by zero in case all play counts for a user are the same
    if max_playcount != min_playcount:
        normalized_playcount = (play_count - min_playcount) / (max_playcount - min_playcount)
    else:
        normalized_playcount = 0  # or another appropriate value, such as 0.5 or 1
    matrix[i] = (music_id, user_id, normalized_playcount)

# Now 'matrix' contains the normalized play counts for all users
```

```python
df = pd.DataFrame(matrix, columns=['track_id', 'user_id', 'normalized_playcount'])
```
