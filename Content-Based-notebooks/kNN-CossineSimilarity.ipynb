```python

#Import all the required packages
import pandas as pd

# Load the movies.csv file into a Pandas dataframe
music_df = pd.read_csv('../data/music_info.csv')

unique_tags = set(tag for tag_list in music_df['tags'] for tag in str(tag_list).split(','))
total_unique_tags = len(unique_tags)

print(f'There are {total_unique_tags} unique tags in the dataset.')

df=music_df.drop(['tags','genre'],axis=1).copy()
df.drop(['track_id','name','artist','spotify_preview_url','spotify_id'],axis=1,inplace=True)
music_df.head(5)
```

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Example function to fit and query the k-NN model
def knn_recommendation(df, n_neighbors=10):
    # Initialize the k-NN model
    knn = NearestNeighbors(n_neighbors=n_neighbors, algorithm='auto', metric='cosine')
    
    # Fit the model on the data
    knn.fit(df)
    
    # Find the k-neighbors for each item in the dataset
    distances, indices = knn.kneighbors(df)
    
    return distances, indices

# Example usage
# Assuming 'df' is your DataFrame with features
n_neighbors = 10  # Number of neighbors to consider
distances, indices = knn_recommendation(df, n_neighbors)

# Example of how to use the results for recommendations
def get_recommendations(item_index, indices, n_recommendations=5):
    # Get the indices of the recommended items
    recommended_indices = indices[item_index, 1:n_recommendations+1]  # Exclude the first item since it's the query item itself
    
    return recommended_indices

# Get recommendations for a specific item
item_index = 0  # Change this to the index of the item you want recommendations for
recommendations = get_recommendations(item_index, indices)
print(f"Recommendations for item {item_index}: {recommendations}")
```
